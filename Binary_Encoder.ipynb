```python
    import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from category_encoders import BinaryEncoder
from sklearn.feature_selection import SelectKBest, chi2

# ML models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("data/cleaned_data.csv") 
df.head()
```

```python
# Separate features and target
X = df.drop(columns=['INJURY'])
y = df['INJURY']
```

```python
X
```

```python
# Apply Binary Encoding to categorical features
encoder = BinaryEncoder()
X_encoded = encoder.fit_transform(X)
```

```python
X_encoded
```

```python
# Chi-Square Feature Selection (Top 30 features)
selector = SelectKBest(score_func=chi2, k=30)
X_selected = selector.fit_transform(X_encoded, y)
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

```python
selector
```

```python
# Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine": SVC(),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42)
}
```

```python
# Train and evaluate
print("✅ Accuracy Using Binary Encoding + Chi-Square Feature Selection:\n")
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print(f"{name}: {acc:.4f}")
```
