```python
# Step 1: Identify groups
invtype_cols = [col for col in data.columns if col.startswith("INVTYPE_")]
impactype_cols = [col for col in data.columns if col.startswith("IMPACTYPE_")]
invage_cols = [col for col in data.columns if col.startswith("INVAGE_")]
acclass_cols = [col for col in data.columns if col.startswith("ACCLASS_")]
```

```python
# Step 2: Select these + target
selected_columns = invtype_cols + impactype_cols + invage_cols + acclass_cols + ['INJURY_NUM']
data = data[selected_columns]
```

```python
# Step 3: Train/test split
X = data.drop('INJURY_NUM', axis=1)
y = data['INJURY_NUM']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Step 4: Train model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Step 5: Save model & column order
joblib.dump(model, "injury_model.pkl")
pd.Series(X.columns).to_csv("model_columns.csv", index=False)
```
